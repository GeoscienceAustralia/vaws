```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pylab

pylab.rcParams['figure.figsize'] = (8.0, 6.0)
pylab.rcParams['figure.dpi'] = 300
pylab.rcParams['font.size'] = 12
pylab.rcParams['legend.numpoints'] = 1
```

All six cases were simulated with regard to re-distribution of damage by tweaking the VAWS.  
The input scenario file is ``carl1_dmg_dist.cfg``.

# Six cases with regard to damage re-distribution

|          |  1  |  2  |  3  |  4  |  5  |  6  |
|:-------- |:---:|:---:|:---:|:---:|:---:|:---:|
|sheeting  |  o  |  x  |  o  |  o  |  x  |  x  |       
|batten    |  o  |  x  |  o  |  x  |  o  |  x  |    
|rafter    |  o  |  x  |  x  |  x  |  x  |  o  |     

# Case 1: All On

```python
df_dmg_case1 = pd.read_csv('./output_case1/house_dmg_idx.csv')
```

```python
ax = df_dmg_case1.plot(x='speed', y='mean', style='bo', label='Case 1')
ax.set_xlabel('Wind speed (m/s)')
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_ylabel('Damage ratio')
ax.grid(1)
ax.legend(loc=2)
```

# Case 2: All off

```python
df_dmg_case2 = pd.read_csv('./output_case2/house_dmg_idx.csv')
```

```python
ax = df_dmg_case1.plot(x='speed', y='mean', style='bo', label='Case 1')
df_dmg_case2.plot(x='speed', y='mean', style='ro', ax = ax, label='Case 2')

ax.set_xlabel('Wind speed (m/s)')
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_ylabel('Damage ratio')
ax.grid(1)
ax.legend(loc=2)
```

```python
for i in range(20):
    df_dmg_case2.plot(x='speed', y=i, style='ro', ax = ax, label='{}_{}'.format('Case 2', i))
    ax.set_xlabel('Wind speed (m/s)')
    ax.set_yticks(np.arange(0, 1.1, 0.1))
    ax.set_ylabel('Damage ratio')
    ax.grid(1)
    ax.legend(loc=2)    
```

```python
df_diff_case1to2 = df_dmg_case1 - df_dmg_case2
df_diff_case1to2['speed'] = df_dmg_case1['speed']

ax = df_diff_case1to2.plot(x='speed', y=range(100), style='bo', legend=False, title='DI difference between Case 1 and Case 2')
ax.set_ylabel('Case1 - Case2')
ax.grid(True)
```

```python
ax = df_diff_case1to2.plot(x='speed', y='mean', style='bo', title='Mean DI difference between Case 1 and Case 2')
ax.set_ylabel('Case1 - Case2')
ax.grid(True)
```
